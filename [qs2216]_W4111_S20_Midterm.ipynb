````python
import logging


def run_q(sql, args=None, fetch=True, cur=None, conn=midterm_conn, commit=True):
    '''
    Helper function to run an SQL statement.

    This is a modification that better supports HW1. An RDBDataTable MUST have a connection specified by
    the connection information. This means that this implementation of run_q MUST NOT try to obtain
    a default connection.

    :param sql: SQL template with placeholders for parameters. Cannot be NULL.
    :param args: Values to pass with statement. May be null.
    :param fetch: Execute a fetch and return data if TRUE.
    :param conn: The database connection to use. This cannot be NULL, unless a cursor is passed.
        DO NOT PASS CURSORS for HW1.
    :param cur: The cursor to use. This is wizard stuff. Do not worry about it for now.
        DO NOT PASS CURSORS for HW1.
    :param commit: This is wizard stuff. Do not worry about it.

    :return: A pair of the form (execute response, fetched data). There will only be fetched data if
        the fetch parameter is True. 'execute response' is the return from the connection.execute, which
        is typically the number of rows effected.
    '''

    cursor_created = False
    connection_created = False

    try:

        if conn is None:
            raise ValueError("In this implementation, conn cannot be None.")

        if cur is None:
            cursor_created = True
            cur = conn.cursor()

        if args is not None:
            log_message = cur.mogrify(sql, args)
        else:
            log_message = sql

        logger.debug("Executing SQL = " + log_message)

        res = cur.execute(sql, args)

        if fetch:
            data = cur.fetchall()
        else:
            data = None

        # Do not ask.
        if commit == True:
            conn.commit()

    except Exception as e:
        if commit == True:
            conn.rollback()
        raise(e)

    return (res, data)
````

    - _productCode_

- Do the following:
    - Create a table _products\_fixed_ that resolves the issue.
    - Load data from the existing _products_ table into the new table.
    
    
- Include the definition of your new table below and execute a sample query.

<hr style="height:1px;">

<u>Answer</u>





Answer goes here.


Table definition.

````python
%%sql
use classicmodels;
CREATE TABLE products_fixed AS
SELECT substr(productCode,1,locate('_',productCode)-1) as scale_prefix,
substr(productCode,locate('_',productCode)+1) as product_number,
substr(productScale,1,locate(':',productScale)-1) as scale_start,
substr(productScale,locate(':',productScale)+1) as scale_suffix,
products.* 
FROM products;
UPDATE products_fixed SET scale_prefix=concat('S',scale_suffix);
UPDATE products_fixed SET product_number=4963 WHERE productCode='S24_3856';

ALTER TABLE `products_fixed` 
CHANGE COLUMN `scale_prefix` `scale_prefix` VARCHAR(15) NOT NULL ,
CHANGE COLUMN `product_number` `product_number` VARCHAR(15) NOT NULL ,
CHANGE COLUMN `productCode` `legacyproductCode` VARCHAR(15) NOT NULL ,
ADD PRIMARY KEY (`scale_prefix`, `product_number`),
DROP COLUMN `scale_suffix`,
ADD COLUMN `newProductCode` VARCHAR(45) GENERATED ALWAYS AS (concat(scale_prefix,'_',product_number)) AFTER `MSRP`;
````

````python
%%sql
use W4111Midterm;
SELECT user_id, last_name, first_name, email, section, 
count(*) as no_of_views,
SEC_TO_TIME(sum(`Minutes Delivered`)*60) as total_viewing_time
FROM students s, video_stats_all v 
WHERE v.`User ID`=s.user_id 
group by s.user_id
order by total_viewing_time DESC
LIMIT 10;
````

````python
%%sql
SELECT DAYOFWEEK(time) as `Day of Week`, HOUR(time) as `hour of day`, count(*) as no_of_views
FROM video_stats_all
WHERE DAYOFWEEK(time) is not NULL
AND HOUR(time) is not NULL
group by `Day of Week`,`hour of day`
order by `Day of Week`,`hour of day` ASC;
````

Put your function and trigger definitions here.

````python
USE `W4111Midterm`
CREATE DEFINER=`root`@`localhost` FUNCTION `create_customer_no`(c_name varchar(128)) RETURNS varchar(12) CHARSET utf8mb4
    READS SQL DATA
BEGIN
	declare result varchar(12);
    declare prefix varchar(3);
	declare count int;
	set prefix=lower(substr(c_name,1,3));
	select count(*) into count from customers where customer_name like(concat(prefix,'%'));
	set count=count+1;
	set result=concat(prefix,'-',count);

RETURN result;
END
````

````python
USE `W4111Midterm`
CREATE DEFINER=`root`@`localhost` TRIGGER `customers_BEFORE_INSERT` 
BEFORE INSERT ON `customers` FOR EACH ROW BEGIN
	declare new_c_id varchar(16);
    set new_c_id=create_customer_no(new.customer_name);
    set new.customer_number=new_c_id;
END
````

````python
USE `W4111Midterm`
CREATE DEFINER=`root`@`localhost` TRIGGER `customers_BEFORE_UPDATE` 
BEFORE UPDATE ON `customers` FOR EACH ROW 
BEGIN
	if new.customer_number!=old.customer_number 
    then set new.customer_number=old.customer_number;
    end if;
END
````
